```python
from keras.datasets import mnist
from autokeras.image.image_supervised import ImageClassifier
```

```python
from autokeras.image.image_supervised import load_image_dataset

x_train, y_train = load_image_dataset(csv_file_path="autokeras_mm/train/label.csv",
                                      images_path="autokeras_mm/train")
print(x_train.shape)
print(y_train.shape)

x_test, y_test = load_image_dataset(csv_file_path="autokeras_mm/test/label.csv",
                                    images_path="autokeras_mm/test")
print(x_test.shape)
print(y_test.shape)
```

```python
clf = ImageClassifier(verbose=True)
clf.fit(x_train, y_train, time_limit=12 * 60 * 60)
clf.final_fit(x_train, y_train, x_test, y_test, retrain=True)
y = clf.evaluate(x_test, y_test)
print(y)
```
